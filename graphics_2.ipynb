```python
from pathlib import Path
import numpy as np
from scipy.stats import iqr
from doa_py import graph
```

## Error a partir de Reverberación

### Resultados para medicion de 0 a 180 ° con saltos de 10 ° ante un sweep de 15 segundos.

```python
methods = ['classic', 'scot', 'roth', 'phat']  # Métodos de estimación a utilizar
for method in methods:    
    graph.generate_heatmap(
        measure="angle_error",
        method_graph=method,
        x_variable="rt60",
        x_label="RT60 [s]",
        x_lim=(1.5, 5.5),
        csv_folder=str(Path("csv_results") / "sweep_rt60"),
        y_label="Error angular [°]",
        y_axis_label="Ángulo de llegada [°]",
        output_dir="img/sweep_rt60",
        graph_mode="interpolated"
    )
```

```python
graph.graficar_error_vs_variable(
    csv_folder="csv_results/sweep_rt60",
    agrupador_x="rt60",
    columna_error="angle_error",
    method_column="method",
    metodos_a_graficar=["CLASSIC", "PHAT", "SCOT", "ROTH"],
    x_label="RT60 [s]",
    y_label="Error angular promedio [°]",
    dispersion_func=iqr,  # ¡IQR robusto!
    output_dir="img/sweep_rt60",
    colores={
        "CLASSIC": "gray",
        "PHAT": "blue",
        "ROTH": "orange",
        "SCOT": "green"
    },
    trazos={
        "CLASSIC": "-.",
        "PHAT": "-",
        "ROTH": "--",
        "SCOT": ":"
    },
    marcadores={
        "CLASSIC": "x",
        "PHAT": "o",
        "ROTH": "s",
        "SCOT": "^"
    }
)
```
