```python
from pathlib import Path
import numpy as np
from scipy.stats import iqr
from doa_py import graph
```

## Error a partir de Reverberaci√≥n

```python
methods = ['classic', 'scot', 'roth', 'phat']  # M√©todos de estimaci√≥n a utilizar
for method in methods:    
    graph.generate_heatmap(
        measure="angle_error",
        method_graph=method,
        x_variable="rt60",
        x_label="RT60 [s]",
        x_lim=(0.5, 5),
        csv_folder=str(Path("csv_results") / "reverb"),
        y_label="Error angular [¬∞]",
        y_axis_label="√Ångulo de llegada [¬∞]",
        output_dir="img/rt60",
        graph_mode="interpolated"
    )
```

```python
graph.graficar_error_vs_variable(
    csv_folder="csv_results/reverb",
    agrupador_x="rt60",
    columna_error="angle_error",
    method_column="method",
    metodos_a_graficar=["CLASSIC", "PHAT", "SCOT", "ROTH"],
    x_label="RT60 [s]",
    y_label="Error angular promedio [¬∞]",
    dispersion_func=iqr,  # ¬°IQR robusto!
    # x_range=(0, 1),
    # y_range=(0, 25),
    output_dir='img/reverb'
)
```

```python
graph.graficar_error_vs_variable(
    csv_folder="csv_results/reverb",
    agrupador_x="rt60",
    columna_error="angle_error",
    method_column="method",
    metodos_a_graficar=["CLASSIC", "PHAT", "SCOT", "ROTH"],
    x_label="RT60 [s]",
    y_label="Error angular promedio [¬∞]",
    dispersion_func=np.std,  # ¬°IQR robusto!
    # x_range=(0, 1),
    # y_range=(0, 25),
    output_dir='img/reverb'
)
```
